```python
import requests as re
from bs4 import BeautifulSoup
import re as reg

# Send GET request to the URL
response = re.get("https://diemthi.vnexpress.net/index/detail/sbd/01000001/year/2023")
soup = BeautifulSoup(response.content, "html.parser") # Parse the HTML content of the response

print(response)
info_detail = str(soup.findAll("p", "o-detail-thisinh__cumthi")) # Find and convert the specific HTML elements to a string
ma_cum_thi = reg.findall('\s+(\d+.+)(</p>)', info_detail)[0][0] # using a regular expression to extract information
print(ma_cum_thi)
```

```python
diem_elements = soup.findAll("div", class_="o-detail-thisinh__diemthi")
diem_element = diem_elements[0]
diem = diem_element.text.strip()
diem = reg.sub(r'\s+', ' ', diem)
pattern = r'(\w+\s\w+)\s(\d+\.\d+)'
matches = reg.findall(pattern, diem)
diem_dict = {subject: float(score) for subject, score in matches}
for subject, score in diem_dict.items():
    print(f"{subject}: {score}")
```

```python
import requests
from bs4 import BeautifulSoup
import re

def get_info_by_id(ID):
    try:
        if int(ID) < 10000000:  # Ensure ID is zero-padded if needed
            ID = f"{ID:08}"

        url = f"https://diemthi.vnexpress.net/index/detail/sbd/{ID}/year/2023"
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception if the request fails (e.g., 404)
        soup = BeautifulSoup(response.content, "html.parser")

        # Extract student's area code
        area_code_element = soup.find("p", class_="o-detail-thisinh__cumthi")
        if not area_code_element:
            return f"Không tìm thấy dữ liệu cho SBD: {ID}"

        area_code = area_code_element.text.split(":")[-1].strip()

        # Extract and format subject scores
        diem_elements = soup.find_all("div", class_="o-detail-thisinh__diemthi")
        diem_dict = {}
        for elem in diem_elements:
            subject, score = elem.text.strip().split(" ")
            diem_dict[subject] = float(score)

        return {'Mã Vùng Thi': area_code,
                'sbd': ID,
                'Điểm': diem_dict}

    except requests.exceptions.RequestException as e:
        return f"Lỗi mạng khi truy cập: {e}"
    except ValueError as e:
        return f"Lỗi định dạng dữ liệu: {e}"
    except Exception as e:
        return f"Lỗi không xác định: {e}"

if __name__ == "__main__":
    Id_01 = 231
    print(get_info_by_id(Id_01))
```

# Browse the website vnexpress.net to retrieve the titles of all articles

```python
import requests
from bs4 import BeautifulSoup

url = 'https://vnexpress.net/'

try:
    response = requests.get(url)
    response.raise_for_status()
except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error: {http_err}")
except requests.exceptions.RequestException as req_err:
    print(f"error sending the request: {req_err}")
else:
    soup = BeautifulSoup(response.text, 'html.parser')
    title_tags = soup.find_all(class_='title-news')
    for title_tag in title_tags:
        title_text = title_tag.get_text().strip()
        print(title_text)
```

#Print the content of each title

```python
import requests
from bs4 import BeautifulSoup

url = 'https://vnexpress.net/'
try:
    response = requests.get(url)
    response.raise_for_status()
except requests.exceptions.HTTPError as http_err:
    print(f"Có lỗi HTTP: {http_err}")
except requests.exceptions.RequestException as req_err:
    print(f"Có lỗi trong quá trình gửi yêu cầu: {req_err}")
else:
    soup = BeautifulSoup(response.text, 'html.parser')
    title_tags = soup.find_all(class_='title-news')
    for title_tag in title_tags:
        title_text = title_tag.get_text().strip()
        link = title_tag.a['href']
        print(title_text)
        sub_response = requests.get(link)
        sub_response.raise_for_status()
        sub_soup = BeautifulSoup(sub_response.text, 'html.parser')
        content_element = sub_soup.find(class_='fck_detail')
        if content_element:
            content = content_element.get_text()
            content = content.replace('\n', '')
            print(f'{content} \n')
        else:
            print("Could not find content with class 'fck_detail'")
        content = content.replace('\n', '')
```
